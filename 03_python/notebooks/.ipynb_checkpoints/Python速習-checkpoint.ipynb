実践課題
絶対値を返す関数absを作る

出力のイメージ
- abs(-1)->1
- abs(-5.2)->5.2
- abs(3)->3

```python
def abs(num):
    if num<0:  #負の値
        num=num*(-1)
    return(num)
```

```python
abs(-1)
```

```python
abs(-5.2)
```

```python
abs(1)
```

## パラメータaの計算

### 傾きaの計算式
$$
a=\frac{\displaystyle\sum_{n=1}^N x_ny_n}{\displaystyle\sum_{n=1}^N x_n^2}
$$

```python
xx=x*x   #*は要素積
```

```python
xy=x*y
```

```python
a=xy.sum()/xx.sum()
```

```python
a
```

```python
float(a)
```

### プロットして確認
$$
\hat{y}=ax
$$

```python
plt.scatter(x,y,label="y")                  #実測値
plt.plot(x,a*x,label='y_hat',color='red')   #予測値
plt.legend()
plt.show()
```

予測値の計算
$$
\begin{align}
y-{\bar y} &= a(x-{\bar y})\\
\rightarrow y &= a(x-{\bar x})+{\bar y}
\end{align}
$$

```python
x_new=40  #40平米の部屋
```

```python
mean=df.mean()
```

```python
mean['x']
```

```python
float(mean['x'])
```

```python
#中心化
xc=x_new-mean['x']
```

```python
xc
```

```python
float(xc)
```

```python
#単回帰分析による予測
yc=a*xc
```

```python
yc
```

```python
float(yc)
```

```python
#元のスケールの予測値
y_hat=a*xc+mean['y']
```

```python
y_hat
```

```python
float(y_hat)
```

### 予測値を計算する関数の作成

```python
def predict(x):
    #定数項
    a=10069.022519284063
    xm=37.62222
    ym=121065.0
    #中心化
    xc=x-xm
    #予測値の計算
    y_hat=a*xc+ym
    #出力
    return y_hat
```

```python
#予測値
predict(30)    #25平米は外挿の範囲になるため、おかしな値が出る (保証できない)
```

```python
predict(10)    #おかしな値 -> 外挿
```

```python
predict(50)
```
