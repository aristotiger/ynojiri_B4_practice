# リスト

```python
# リストの生成
a = [1,2,3,4]

print( a )
```

```python
# 5～100までのリストの生成
a = list(range(5, 101))

print( a )
```

```python
# [0.5, 0.6, 0.7, ..., 10.0]というリストの生成
a = [ i/10.0 for i in range(5, 101)]

print( a )
```

```python
# リストの要素数
a = list(range(546))

print( len(a) )
```

```python
# 2つのリストの連結
a = list(range(10))
b = list(range(10, 20))
c = a + b

print( c )
```

```python
# 名前のリスト["Taro", "Hanako", "Jiro"]と，テストの点数のリスト[95, 80, 90]がある時，
# 要素を連結したリスト[("Taro", 95), ("Hanako", 80), ("Jiro", 90)]の生成
names = ["Taro", "Hanako", "Jiro"]
points = [95, 80, 90]
pairs = list(zip(names, points))

print( pairs )
```

```python
# [("Taro", 95), ("Hanako", 80), ("Jiro", 90)]の要素をfor文で一つづつ表示
names = ["Taro", "Hanako", "Jiro"]
points = [95, 80, 90]
pairs = zip(names, points)

for name, point in pairs:
    print( "Name: ", name )
    print( "Point: ", point )
    print( "------" )
```

```python
# 適当なリストの50番目以降の値を取り出す
a = list(range(74,246))
b = a[50:]

print( b )
```

```python
#  適当なリストから50番目以降の6個の値を取り出す
a = list(range(74,246))
b = a[50:56]

print( b )
```

```python
#  適当なリストから50番目以降の値を7つ置きに値を取り出す
a = list(range(74,246))
b = a[50::7]

print( b )
```

```python
# [0, 1, 2, ..., 99]のリストを順序を逆にする
a = list(range(100))
b = a[::-1]

print( b )
```

```python
# リストの後ろから3つ目の要素を取り出す
a = [5, 7, 1, 3, 2]
b = a[-3]

print( b )
```
